```python
# For data manipulation

import pandas as pd

import numpy as np 

# machine learning classification

from sklearn.svm import SVC
from sklearn.svm import NuSVC

from sklearn.metrics import scorer

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# To plot
%matplotlib inline
import matplotlib.pyplot as plt

import seaborn

# To fetch data
```

```python
import glob
import pandas as pd
pattern='*.csv'
hong = glob.glob(pattern)
len(hong)
```

```python
hkong = pd.read_csv(hong[0])
hkong = hkong.drop('Adj Close', axis=1)
hkong.columns = ['Date', 'Open' + '_'+str(0), 'High'+'_'+str(0), 'Low' +'_'+str(0), 'Close'+ '_'+str(0), 'Volume'+ '_'+ str(0)]
hkong = hkong.fillna(0)
#hkong['Target'] = np.where(hkong['Close'].shift(-1) > hkong['Close'],int(1),int(-1))
hkong.shape
```

```python
hkong.head()
```

```python
#run only once
for i in range(1,len(hong)):
    df = pd.read_csv(hong[i]) 
    
    #if 'Adj Close' not in df.columns:
    #    df['Adj Close'] = 0
    #if 'Volume' not in df.columns:
      #  df['Volume'] = 0
    
    df = df.drop('Adj Close', axis=1)  
    df = df.fillna(0)
    #df['Target'] = np.where(df['Close'].shift(-1) > df['Close'], int(1),int(-1))
    
    df.columns = ['Date', 'Open' + '_'+str(i), 'High'+'_'+str(i), 'Low' +'_'+str(i), 'Close'+'_'+str(i), 'Volume'+'_'+str(i)]
    
    hkong = pd.merge(hkong, df, how='outer', on='Date')
hkong = hkong.fillna(int(0))
hkong.shape
```

```python
hkong.head()
```

```python
hkong.isnull().sum()
```

```python
for i in range(30):
    hkong['Target'+'_'+str(i)] = np.where(hkong['Close'+'_'+str(i)].shift(-1) > hkong['Close'+'_'+str(i)],1,-1)
    
hkong.shape
```

```python
hkong.head()
```

```python
hkong.index = hkong.Date
hkong = hkong.drop('Date', axis=1)
hkong.shape
```

```python
hkong.head()
```

```python
hkong.to_csv('E:/hkong.csv')
```

```python
target=hkong.iloc[:, 290:].columns
target
```

```python
X = hkong.iloc[:, :290]
X.shape
```

```python
scores = []
test_acc=[]

for col in target:
    y = hkong[col]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    

    classifier = LogisticRegression()

    cv_score = np.mean(cross_val_score(classifier, X_test, y_test, cv=10, scoring='roc_auc'))
    scores.append(cv_score)
    print('CV score for class {} is {}'.format(col, cv_score))

    classifier.fit(X_train, y_train)
    accuracy_train = accuracy_score(y_train, classifier.predict(X_train))

    accuracy_test = accuracy_score(y_test, classifier.predict(X_test))
    #print('\nTrain Accuracy:{: .2f}%'.format(accuracy_train*100))

    #print('Test Accuracy:{: .2f}%'.format(accuracy_test*100))
    test_acc.append(accuracy_test)
    #submission[class_name] = classifier.predict_proba(test_features)[:, 1]

print('Total CV score is {}'.format(np.mean(scores)))
print('Mean Accuracy score is {}%'.format(np.mean(test_acc)*100))
```
