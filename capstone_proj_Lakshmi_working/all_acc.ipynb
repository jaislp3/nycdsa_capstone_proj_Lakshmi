```python
import pandas as pd
import numpy as np
```

```python
span1_0p5 = pd.read_csv('E:/hangseng/span1_0p5.csv', index_col='Stock Components')
span5_0p5 = pd.read_csv('E:/hangseng/span5_0p5.csv', index_col='Stock Components')
span10_0p5 = pd.read_csv('E:/hangseng/span10_0p5.csv', index_col='Stock Components')
span1_0p1 = pd.read_csv('E:/hangseng/span1_0p1.csv', index_col='Stock Components')
span1_0p0 = pd.read_csv('E:/hangseng/span1_0p0.csv', index_col='Stock Components')
```

```python
all_acc = pd.concat([span1_0p5, span5_0p5,span10_0p5, span1_0p1, span1_0p0], axis=1)
all_acc.reset_index(inplace=True)
```

```python
all_acc.to_csv('E:/hangseng/all_acc.csv', index=False)
```
